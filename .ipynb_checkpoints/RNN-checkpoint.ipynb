```python
import torch
import torch.nn as nn
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
from Utils import *
```

```python
import torch

def save_checkpoint(model,optimizer,filepath):
    state = {
    'state_dict': model.state_dict(),
    'optimizer': optimizer.state_dict()
    }
    torch.save(state, filepath)

def load_checkpoint(model,optimizer,filepath):
    # "lambda" allows to load the model on cpu in case it is saved on gpu
    state = torch.load(filepath,lambda storage, loc: storage)
    model.load_state_dict(state['state_dict'])
    optimizer.load_state_dict(state['optimizer'])
    return model,optimizer
```

```python
TRAIN_START_PERIOD = '1990-01-01'
TRAIN_END_PERIOD = '1995-01-01'

VAL_START_PERIOD = '1995-01-01'
VAL_END_PERIOD = '1997-01-01'

TEST_START_PERIOD = '1997-01-01'
TEST_END_PERIOD = '1999-01-01'
```

# Dataset Creation

I implemented the RNN model using pytorch.

Pytorch requires to implement dataset class for custom datasets.

In create_data_loaders function I used the FinanceDataset class to create train, validation and test datasets.

The features used in the data is the current values of interest rates (['3 Mo', '1 Yr', '5 Yr', '10 Yr', '30 Yr']) and the  spread.

```python
from torch.utils.data import Dataset, DataLoader

class FinanceDataset(Dataset):
    def __init__(self, start_date, end_date, asset_to_forecast, time_to_forecast, scaler):
        self.finance_data_df = create_data_df(use='dataset', start_date=start_date, end_date=end_date)
        
        self.fitted_finance_data_df = pd.DataFrame(data=scaler.transform(self.finance_data_df),
                                                   columns=self.finance_data_df.columns)
                
        self.labels_df = self.fitted_finance_data_df.shift(-time_to_forecast)[asset_to_forecast].dropna()
        self.fitted_finance_data_df = self.fitted_finance_data_df.loc[self.labels_df.index[0]:self.labels_df.index[-1]]

        self.data_tensor = torch.from_numpy(self.fitted_finance_data_df.values)
        self.labels_tensor = torch.from_numpy(self.labels_df.values)
       
    def __len__(self):
        return self.fitted_finance_data_df.shape[0]
    
    def __getitem__(self, idx):
        return self.data_tensor[idx], self.labels_tensor[idx]
```

```python
def create_data_loaders(period, time_to_predict, batch_size, scaler):
    train_finance_dataset = FinanceDataset(TRAIN_START_PERIOD, TRAIN_END_PERIOD, period, time_to_predict, scaler)
    val_finance_dataset = FinanceDataset(VAL_START_PERIOD, VAL_END_PERIOD, period, time_to_predict, scaler)
    test_finance_dataset = FinanceDataset(TEST_START_PERIOD, TEST_END_PERIOD, period, time_to_predict, scaler)
    
    train_data_loader = DataLoader(train_finance_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
    val_data_loader = DataLoader(val_finance_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
    test_data_loader = DataLoader(test_finance_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
    
    return train_data_loader, val_data_loader, test_data_loader
```

# RNN Module

```python
class RNN(nn.Module):
    def __init__(self, data_dim, output_size, hidden_dim, n_layers):
        super(RNN, self).__init__()
        
        self.hidden_dim=hidden_dim

        # define an RNN with specified parameters
        # batch_first means that the first dim of the input and output will be the batch_size
        self.rnn = nn.RNN(data_dim, hidden_dim, n_layers, batch_first=True)
        
        # last, fully-connected layer
        self.fc = nn.Linear(hidden_dim, output_size)

        self.dropout = nn.Dropout(0.9)
        
    def forward(self, x, hidden):
        # x (batch_size, seq_length, input_size)
        # hidden (n_layers, batch_size, hidden_dim)
        # r_out (batch_size, time_step, hidden_size)
        batch_size = x.size(0)
        
        # get RNN outputs
        r_out, hidden = self.rnn(x, hidden)
        # shape output to be (batch_size*seq_length, hidden_dim)
        r_out = r_out.view(-1, self.hidden_dim)  
        
        # get final output 
        output = self.dropout(self.fc(r_out))
        
        return output, hidden
```

# Train, Validate and Run functions

Train function used to train the model.
We save the model that get the lowerest loss value on the validation set.

```python
PATH = 'rnn.pkl'

def train(rnn, optimizer, criterion, train_data_loader, val_data_loader):
    hidden = None      
    min_loss = 100
    
    rnn.train()
    
    train_loss_list=list()
    val_loss_list=list()
    for batch_i, data in enumerate(train_data_loader):
        optimizer.zero_grad()

        inputs, labels = data

        inputs = inputs.unsqueeze(0)
        inputs = inputs.float()
        labels = labels.float()

        prediction, hidden = rnn(inputs, hidden)
        hidden = hidden.data

        labels = labels.unsqueeze(1)
        # calculate the loss
        loss = criterion(prediction, labels)
        # zero gradients
        # perform backprop and update weights
        loss.backward()
        optimizer.step()

        train_loss = loss.item()
        val_loss, _, _ = validate(rnn, optimizer, criterion, val_data_loader)
        train_loss_list.append(train_loss)
        val_loss_list.append(val_loss)

#             print('batch: {}, Train_Loss: {} '.format(batch_i, train_loss))
#             print('batch: {}, Val_Loss: {} '.format(batch_i, val_loss))
        if val_loss < min_loss:
            min_loss = val_loss
            save_checkpoint(rnn,optimizer,PATH)
                
    return train_loss_list, val_loss_list   
```

Validate function is used to evaludate the rnn model performence on validation and test datasets.

```python
def validate(trained_rnn, optimizer, criterion, data_loader):
    prediction_list = list()    
    labels_list = list()
    
    hidden = None      
    trained_rnn.eval()
    current_loss=0
    
    for batch_i, data in enumerate(data_loader):
        inputs, labels = data
        batch_size = inputs.size(0)
        inputs = inputs.unsqueeze(0)
        inputs = inputs.float()
        labels = labels.float()
        labels = labels.unsqueeze(1)

        with torch.no_grad():
            prediction, hidden = trained_rnn(inputs, hidden)
            loss = criterion(prediction, labels)
            current_loss += loss.item()

        prediction_list.extend(prediction.numpy().tolist())
        labels_list.extend(labels.numpy().tolist())

    current_loss = current_loss/(batch_size*batch_i)
    return current_loss, prediction_list, labels_list


        
```

Run function is used to run an experiment of prediction future price of specific time series.

```python
import matplotlib.pyplot as plt
import math  
from sklearn.metrics import mean_squared_error

def inverse_transfrom(data, mean, std):
    data = np.asarray(data)*std+mean
    return data.tolist()

def run(period, time_to_predict, hidden_dim, n_layers, lr, batch_size, scaler, scaling_dict):
    train_data_loader, val_data_loader, test_data_loader = create_data_loaders(period, time_to_predict, batch_size, scaler)
    
    dataiter = iter(train_data_loader)
    samples, _ = dataiter.next()
    input_size=samples.size(1) 
    output_size=1


    rnn = RNN(input_size, output_size, hidden_dim, n_layers)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(rnn.parameters(), lr=lr)
    
    train_loss_list, val_loss_list = train(rnn, optimizer, criterion, train_data_loader, val_data_loader)
    trained_rnn, optimizer = load_checkpoint(rnn,optimizer, PATH)
    val_loss, val_prediction_list, val_labels_list = validate(trained_rnn, optimizer, criterion, val_data_loader)
    
    mean = scaler.mean_[scaling_dict[period]]
    std = math.sqrt(scaler.var_[scaling_dict[period]])
    
    val_prediction_list = inverse_transfrom(val_prediction_list, mean, std)
    val_labels_list = inverse_transfrom(val_labels_list, mean, std)
    
    print('Train \ Validation Loss')
    plt.figure()
    plt.plot(train_loss_list, 'b', label='train_loss')
    plt.plot(val_loss_list, 'r', label='validation_loss')
    plt.legend(loc='best')
    plt.show()
    
    print('Predictions on validation set')
    plt.figure()
    plt.plot(val_prediction_list, 'r', label='prediction')
    plt.plot(val_labels_list, 'b', label='actual')
    plt.legend(loc='best')
    plt.show()
    
    _, test_prediction_list, test_labels_list = validate(trained_rnn, optimizer, criterion, test_data_loader)    
    
    test_prediction_list = inverse_transfrom(test_prediction_list, mean, std)
    test_labels_list = inverse_transfrom(test_labels_list, mean, std)
    
    print('Prediction on test set')
    plt.figure()
    plt.plot(test_prediction_list, 'r', label='prediction')
    plt.plot(test_labels_list, 'b', label='actual')
    plt.legend(loc='best')
    plt.show()
    
    test_loss = mean_squared_error(test_labels_list, test_prediction_list)
    print('MSE loss on test set: {test_loss}'.format(test_loss=test_loss))
    
```

```python
PERIODS = ['3 Mo', '1 Yr', '5 Yr', '10 Yr', '30 Yr']
PERIODS = PERIODS + ['BAA10Y']

data_df = create_data_df(use='dataset', start_date='1990-01-01', end_date='2000-01-01')
month_shift = data_df.loc[VAL_START_PERIOD:'1995-02-01'].shape[0]-1
year_shift = data_df.loc[VAL_START_PERIOD:'1996-01-01'].shape[0]
time_to_predict = {'day': 1,
                   'month': month_shift,
                   'year': year_shift}
```

# Results

I ran the each test for 1 epoch to avoid the vanishing gradient phenomena in RNN.

We must normalize the values in the data since there it is possible that the model deal with values at validation and test sets that it did not encountered before. Without normalizing the data the model will not be able to deal with these values and the prediction in the validation and test sets will be a constant number that will yield the lowest MSE loss in the validation set.

I normalized the data using the available data only, meaning the train and validation sets.

The first plot of the loss is refers to MSE loss on the normalized data.

The MSE loss mentioned on the test set is on non-normalized values.

In most experiments there is a point in time when the loss of the validation set is approximately. We can save the model that yield this result and use it on the test set.

In some cases there are pretty accurate predictions on the test set and on other cases less accurate predictions.

I used the same RNN model with same hyper parameters for all experiments.

# Spread

The model's prediction does not look as accurate as the prediction gained for the different interest rates when trying to predict a day or a week ahead.

For day ahead prediction:

1. The train loss is not stable (not monotonically decreasing) and the validation prediction is almost a constant value (~1.6 - 1.7. It does not manage to predict the spread start value and the big increase (from 1.3 to 1.9 in around a year). The model succeed to predict that there is an increase in the spread value but it does not manage to predict the magnitude of the increase correctly. This rapid increase in spread value (from 1.8 to 2.8 in around 2 months) might meen that there was a major event which the model failed to predict.

2. The month ahead prediction is approxiamtely constant and does not manage to predict the spread.

3. The year ahead prediction managed to predict the decreasing trend in the validation set, but it predicted the decreasing trend also in the test set when actually the is increasing.

For summary, the model did not managed to predict accurately on the spread data.
